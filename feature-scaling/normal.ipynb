### Normal Feature Scaling

From 0 to 1


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([300, 4560, 500, 2300, 800, 780, 1200, 4500, 3000])
```

```python
plt.scatter(x_train, x_train, marker="x", c="r")
plt.show()
```

```python
max_element = x_train.max()

x_train = x_train / max_element
```
