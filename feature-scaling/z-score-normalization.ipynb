### Z Score Normalization

for -3 to 3

X_j = (X_j - mean) / std_deviation

![Image](https://tiddi.kunalsin9h.com/ZJZh9)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1000, 2300, 1200, 2000, 1500])
y_train = np.array([1, 23, 12, 2, 15])
```

```python
plt.scatter(x_train, y_train, marker="H", c="blue")
plt.show()
```

```python
def zscore_normalization(x: np.ndarray):
    mean = x.mean()
    std_deviation = x.std()

    return (x - mean) / std_deviation
```

```python
x_train = zscore_normalization(x_train)
y_train = zscore_normalization(y_train)
```

```python
plt.scatter(x_train, y_train, marker="H", c="green")
plt.show()
```
