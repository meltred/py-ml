### Mean Normalization

From -1 to +1 (Desired Values to be in each Features)

x_j = (x_j - mean) / max - min

![PPT](https://tiddi.kunalsin9h.com/rAi11)


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("Solarize_Light2")
```

```python
x_train = np.array([300, 4560, 500, 2300, 800, 780, 1200, 4500, 3000])
```

```python
plt.scatter(x_train, x_train, marker="x", c="r")
plt.show()
```

```python
mean = x_train.mean()
max = x_train.max()
min = x_train.min()

range = max - min
```

```python
x_train = (x_train - mean) / range
```

```python
plt.scatter(x_train, x_train, marker="H", c="blue")
plt.show()
```
